```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Load dataset
df = pd.read_csv("Mall_Customers.csv")
```

```python
# Step 1: Drop CustomerID (not needed for clustering)
df = df.drop("CustomerID", axis=1)
```

```python
# Step 2: Convert 'Genre' to numerical
df['Genre'] = df['Genre'].map({'Male': 0, 'Female': 1})
```

```python
# Step 3: Check for missing values
print("Missing Values:\n", df.isnull().sum())
```

```python
# Step 4: Exploratory Data Analysis (EDA)
print("\nBasic Statistics:\n", df.describe())
```

```python
# Pairplot to explore feature relationships
sns.pairplot(df, diag_kind='kde')
plt.suptitle('Pairplot of Customer Features', y=1.02)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 5: Scale the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
# Create a DataFrame for scaled data
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
print("\nScaled Data Sample:")
print(df_scaled.head())
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
from sklearn.decomposition import PCA
```

```python
# Find optimal number of clusters using Elbow Method
inertia = []
range_n = range(1, 11)
for k in range_n:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Plot Elbow Curve
plt.figure(figsize=(6,4))
plt.plot(range_n, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Silhouette Score
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df_scaled)
    score = silhouette_score(df_scaled, labels)
    print(f'Silhouette Score for k={k}: {score:.3f}')
```

```python
# Apply K-Means with optimal k (let's assume k=5 based on Elbow/Silhouette)
k_optimal = 5
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
clusters = kmeans.fit_predict(df_scaled)
df['Cluster'] = clusters
```

```python
# visualize Clusters using PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(df_scaled)
df['PCA1'] = reduced[:, 0]
df['PCA2'] = reduced[:, 1]
```

```python
# Scatter plot of clusters
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set2', s=80)
plt.title('Customer Segments Visualized using PCA')
plt.show()
```

```python
# View average values for each cluster
cluster_profiles = df.groupby('Cluster').mean(numeric_only=True)
print("Cluster Profiles:\n", cluster_profiles)
```

```python
# Add Gender back for profiling (optional)
df['Gender'] = df['Genre'].map({0: 'Male', 1: 'Female'})
```

```python
# Count of each gender per cluster
gender_distribution = df.groupby(['Cluster', 'Gender']).size().unstack()
print("\nGender Distribution per Cluster:\n", gender_distribution)
```

```python
# Barplot of average income and spending score by cluster
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.barplot(x='Cluster', y='Annual Income (k$)', data=df, ax=ax[0], palette='Set2')
sns.barplot(x='Cluster', y='Spending Score (1-100)', data=df, ax=ax[1], palette='Set3')
ax[0].set_title("Average Annual Income by Cluster")
ax[1].set_title("Average Spending Score by Cluster")
plt.tight_layout()
plt.show()
```
